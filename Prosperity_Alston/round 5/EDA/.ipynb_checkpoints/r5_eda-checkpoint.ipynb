# EDA for round 5 trading data



## Assumptions that need the data to speak out?

- for certiain trader names, they have better performace than others(i.e informed traders), so we copy trading their trades and stop doing counterpart trades with them
- for certain trader species, they have better performace than others(i.e informed traders), so we copy trading their trades and stop doing counterpart trades with them
- for certain trade species, they have information for certain underlying that is not available to others, so we copy trading their trades and stop doing counterpart trades with them when we are trading certain goods

The performance can be measured by their volumed average price to sell a certain good - the volume weighted average price of the same good in the market.

```python
# load libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
trades = pd.read_csv('trades_round_5.csv',sep=';')
```

```python
trades.head()
```

```python
def plot_market_part(trade_df):
    buy = trade_df['buyer'].value_counts()
    sell = trade_df['seller'].value_counts()

    plt.figure(figsize=(10, 5))
    plt.title('Market trades value count by participants')
    plt.bar(buy.index, buy.values, label='Buyers',alpha = 0.5, color = 'red')
    plt.bar(sell.index, sell.values, label='Sellers',alpha = 0.5, color = 'blue')
    # attach the value labels to the bars
    for i in range(len(buy)):
        plt.text(buy.index[i], buy.values[i], buy.values[i], ha='center', va='bottom',color='red')
    for i in range(len(sell)):
        plt.text(sell.index[i], sell.values[i], sell.values[i], ha='center', va='bottom',color='blue')
    plt.show()
```

```python

plot_market_part(trades)
```

```python
species_dict = {"Peter": "P", "Gary": "G", "Mitch": "M", "Penelope": "P",
                "Omar": "O", "Camilla": "C", "Caesar": "C", "Giulia": "G", 
                "Mabelthe": "M", "Charlie": "C", "Pablo": "P", "Olivia": "O", "Orson": "O", "Casey": "C","George":"G","Mya":"M","Max":"M","Paris":"P","Gina":"G","Olga":"O"}
```

```python
def generate_participant_species(trade_df):
    trade_df['buyer_species'] = trade_df['buyer'].map(species_dict)
    trade_df['seller_species'] = trade_df['seller'].map(species_dict)
    return trade_df.copy()
```

```python
trades = trades.pipe(generate_participant_species)
```

```python
def plot_market_part_kind(trade_df,pct = 1):
    buy = trade_df['buyer_species'].value_counts()
    sell = trade_df['seller_species'].value_counts()

    plt.figure(figsize=(10, 5))
    if pct ==0:
        plt.title('Market trades value count by participants species')
    else:
        plt.title("Market participant species percentage")
    plt.bar(buy.index, buy.values, label='Buyers',alpha = 0.5, color = 'red')
    plt.bar(sell.index, sell.values, label='Sellers',alpha = 0.5, color = 'blue')
    # attach the value labels to the bars
    # attach the percentage labels to the bars
    if pct == 0:
        for i in range(len(buy)):
            plt.text(buy.index[i], buy.values[i], buy.values[i], ha='right', va='top',color='red')
        for i in range(len(sell)):
            plt.text(sell.index[i], sell.values[i], sell.values[i], ha='right', va='bottom',color='blue')
    else:
        for i in range(len(buy)):
            plt.text(buy.index[i], buy.values[i], str(round(buy.values[i]/buy.values.sum()*100,2))+'%', ha='left', va='top',color='red')
        for i  in range(len(sell)):
            plt.text(sell.index[i], sell.values[i], str(round(sell.values[i]/sell.values.sum()*100,2))+'%', ha='left', va='bottom',color='blue')
    
    plt.show()
```

```python
trades.pipe(plot_market_part_kind)
trades.pipe(plot_market_part_kind,0)
```

```python
name_ls = list(species_dict.keys())
```

```python
for name in name_ls:
    # if the name is not in the list of participants, skip
    if name not in trades['buyer'].unique() and name not in trades['seller'].unique():
        continue
    else:
        print(name)
        plt.figure(figsize=(9, 6))
        plt.title(name)
        df = trades[(trades['buyer'] == name) | (trades['seller'] == name)]
        
        df['cashflow'] = (-df['price']*(df['buyer'] == name).astype(int) +
                        df['price']*(df['seller'] == name).astype(int))*df['quantity']
        # for each of the ticker symbols, plot the cumulative cashflow in  a single plot
        grouped = df.groupby(['symbol'])
        for key, group in grouped:
            if key == 'COCONUTS' or key == 'PINA_COLADAS':
                continue
            else:
                plt.plot(group['timestamp'], group['cashflow'].cumsum(), label=key)

        # plot the cumulative cashflow of the entire portfolio
        plt.plot(df['timestamp'], df['cashflow'].cumsum(), label='All')


        # if the name trade coconut and pina colada, plot the cumulative cashflow of the two
        if 'COCONUTS' in df['symbol'].unique() and 'PINA_COLADAS' in df['symbol'].unique():
            plt.plot(df[(df['symbol']=='COCONUTS')|(df['symbol']=='PINA_COLADAS')]['timestamp'], df[(df['symbol']=='COCONUTS')|(df['symbol']=='PINA_COLADAS')]['cashflow'].cumsum(), label='C & PC pair 1')

        # if the trader trade PICNIC_BASKET,DIP,BAGUETTE and UKULELE at the sampe time, plot the cumulative cashflow of the two
        if 'PICNIC_BASKET' in df['symbol'].unique() and 'DIP' in df['symbol'].unique() and 'BAGUETTE' in df['symbol'].unique() and 'UKULELE' in df['symbol'].unique():
            plt.plot(df[(df['symbol']=='PICNIC_BASKET')|(df['symbol']=='DIP')|(df['symbol']=='BAGUETTE')|(df['symbol']=='UKULELE')]['timestamp'], df[(df['symbol']=='PICNIC_BASKET')|(df['symbol']=='DIP')|(df['symbol']=='BAGUETTE')|(df['symbol']=='UKULELE')]['cashflow'].cumsum(), label='PB & D & U & B pair 2')

        plt.legend()
        plt.show()
```
