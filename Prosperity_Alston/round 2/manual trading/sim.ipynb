```python
# load library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# run a simulation 
# generate 1500 int r.v. of uniform distribution of 9000 to 11000 

# set seed
np.random.seed(123)

# generate 1500 int r.v. of uniform distribution of 9000 to 11000
# 1500 is the number of active traders
# 9000 is the lowest price
# 11000 is the highest price

rv = np.random.randint(9000, 11000, 1500)

# calculate the median of the 1500 r.v.
median = np.median(rv)
# calculate the average of r.v. below the median
avg_below = np.mean(rv[rv < median])

print('The median is', median)
print('The average of r.v. below the median is', avg_below)
```
