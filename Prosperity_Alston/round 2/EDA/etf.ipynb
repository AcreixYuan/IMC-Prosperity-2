```python
import numpy as np
import pandas as pd
```

```python
etf_df = pd.read_csv('best_info.csv')
print(etf_df)
```

```python

#fair price = 1000
coco_pina_df = pd.DataFrame(columns = ['mid_price','best_bid','best_ask','best_bid_volume','best_ask_volume','fair_price'])
coco_pina_df['mid_price'] = 2*(etf_df['COCONUTS_best_ask_price']+etf_df['COCONUTS_best_bid_price'])/2 - (etf_df['PINA_COLADAS_best_ask_price']+etf_df['PINA_COLADAS_best_bid_price'])/2
#best bid = 2coconutbest_bid-1* pina_colada*best_ask
coco_pina_df['best_bid'] = 2*(etf_df['COCONUTS_best_bid_price']) - (etf_df['PINA_COLADAS_best_ask_price'])
#best ask = 2coconutbest_ask-1* pina_colada*best_bid
coco_pina_df['best_ask'] = 2*(etf_df['COCONUTS_best_ask_price']) - (etf_df['PINA_COLADAS_best_bid_price'])

#best bid volume = 2coconutbest_bid_volume-1* pina_colada*best_ask_volume
coco_pina_df['best_bid_volume'] = etf_df.apply(lambda row: min(2*row['COCONUTS_best_bid_volume'],row['PINA_COLADAS_best_ask_volume']),axis=1)

#best ask volume = 2coconutbest_ask_volume-1* pina_colada*best_bid_volume
coco_pina_df['best_ask_volume'] = etf_df.apply(lambda row: min(2*row['COCONUTS_best_ask_volume'],row['PINA_COLADAS_best_bid_volume']),axis=1)
coco_pina_df['fair_price'] = 1000
coco_pina_df
```

```python
coco_pina_df.to_csv('virtual_etf.csv',index = False)
```
