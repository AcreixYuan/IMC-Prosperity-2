```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pnl_linear = pd.read_excel('weighted_4prod_linear.xlsx')
pnl_ratio = pd.read_excel('weighted_4prod_ratio.xlsx')
```

```python
pnl_pearls_linear = pd.Series(pnl_linear[pnl_linear['product'] == 'PEARLS']['profit_and_loss']).reset_index(drop=True)
pnl_bananas_linear = pd.Series(pnl_linear[pnl_linear['product'] == 'BANANAS']['profit_and_loss']).reset_index(drop=True)
pnl_coconut_linear = pd.Series(pnl_linear[pnl_linear['product'] == 'COCONUTS']['profit_and_loss']).reset_index(drop=True)
pnl_pina_linear = pd.Series(pnl_linear[pnl_linear['product'] == 'PINA_COLADAS']['profit_and_loss']).reset_index(drop=True)

pnl_pearls_ratio = pd.Series(pnl_ratio[pnl_ratio['product'] == 'PEARLS']['profit_and_loss']).reset_index(drop=True)
pnl_bananas_ratio = pd.Series(pnl_ratio[pnl_ratio['product'] == 'BANANAS']['profit_and_loss']).reset_index(drop=True)
pnl_coconut_ratio = pd.Series(pnl_ratio[pnl_ratio['product'] == 'COCONUTS']['profit_and_loss']).reset_index(drop=True)
pnl_pina_ratio = pd.Series(pnl_ratio[pnl_ratio['product'] == 'PINA_COLADAS']['profit_and_loss']).reset_index(drop=True)

```

```python
plt.plot(pnl_pearls_linear, label='PEARLS')
plt.plot(pnl_bananas_linear, label='BANANAS')
plt.plot(pnl_coconut_linear+pnl_pina_linear, label='COCONUTS+PINA_COLADAS')
plt.legend()
plt.show()
```

```python
plt.plot(pnl_pearls_ratio, label='PEARLS')
plt.plot(pnl_bananas_ratio, label='BANANAS')
plt.plot(pnl_coconut_ratio+pnl_pina_ratio, label='COCONUTS+PINA_COLADAS')
plt.legend()
plt.show()
```

```python
pair_pnl_linear = pnl_pina_linear+pnl_coconut_linear
```

```python
#calculate sharpe ratio
def sharpe_ratio(returns, periods=252):
    return np.sqrt(periods) * (returns.mean() / returns.std())
```

```python
sharpe_ratio(pair_pnl_linear.diff())
```

```python
pair_pnl_ratio = pnl_pina_ratio+pnl_coconut_ratio
```
