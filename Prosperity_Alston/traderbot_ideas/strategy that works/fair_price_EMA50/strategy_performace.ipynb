```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
strat_report = pd.read_csv("pearl_ema_price.csv",sep = ';')
```

```python
strat_report[strat_report['product'] == 'PEARLS'].plot(x = 'timestamp', y = ['profit_and_loss'])
plt.xlabel('tick')
plt.ylabel('Dollar PnL')
plt.legend(loc='upper left', bbox_to_anchor=(1.0, 1.0))
plt.show()
```

```python
strat_report[strat_report['product'] == 'PEARLS']['profit_and_loss']
```

```python
pnl = strat_report[strat_report['product'] == 'PEARLS']['profit_and_loss'].diff(1)
```

```python
pnl.plot(x = 'timestamp', y = ['profit_and_loss'],label = 'PnL')
plt.legend()
```

```python
def cal_sharpe_ratio(pnl):
    return np.sqrt(252) * pnl.mean() / pnl.std()
```

```python
cal_sharpe_ratio(pnl)
```
